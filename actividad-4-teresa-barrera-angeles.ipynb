```python
### CONTINUA NO PERIODICA
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de tiempo t
t = np.linspace(-1, 2, 1000)  # Genera 1000 puntos en el rango de -1 a 2

# Definir la función x(t)
T = 1  # Período
x = np.zeros_like(t)
x[(t >= 0) & (t <= T)] = 1

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar la función
ax.plot(t, x)

# Configurar los ejes y etiquetas
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Gráfica de x(t/T)')

# Mostrar la gráfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Definir la función continua
def f(x):
    return np.sin(x) + np.exp(-x)
# Definir los parámetros para la representación de Fourier
N = 1000  # Número de puntos en la representación de Fourier
T = 2 * np.pi  # Período de la señal periódica
# Calcular los coeficientes de Fourier
x = np.linspace(-10, 10, N)
y = f(x)
frequencies = np.fft.fftfreq(N, d=x[1]-x[0])
coeficients = np.fft.fft(y)
# Graficar la función original
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Función Continua')
plt.xlabel('x')
plt.ylabel('f(x)')
# Graficar la representación de Fourier
plt.subplot(1, 2, 2)
plt.stem(frequencies, np.abs(coeficients))
plt.title('Representación de Fourier')
plt.xlabel('Frecuencia')
plt.ylabel('Coeficiente')
plt.xlim(-10, 10)  # Limitar el rango de frecuencias
plt.ylim(0, np.max(np.abs(coeficients)) * 1.1)  # Ajustar el rango del eje y
plt.grid(True)
# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

$ f(t) = 3sin(2πt/T) + 2cos(4πt/T) $ 
Donde T es el período de la señal.

Para calcular los coeficientes de Fourier de esta señal
##### Paso 1: Calcula el valor medio (componente DC):
$a_0 = (1/T) * ∫[0,T] f(t) dt$
#### Para la señal dada, el valor medio es:
$a_0 = (1/T) * ∫[0,T] [3sin(2πt/T) + 2cos(4πt/T)] dt$
##### Paso 2: Calcula los coeficientes de Fourier (componentes armónicos):
$a_n = (2/T) * ∫[0,T] f(t) * cos(2πnt/T) dt
b_n = (2/T) * ∫[0,T] f(t) * sin(2πnt/T) dt$
##### Donde $n$ es un número entero positivo. Para la señal dada, los coeficientes de Fourier son:
$a_n = (2/T) * ∫[0,T] [3sin(2πt/T) + 2cos(4πt/T)] * cos(2πnt/T) dt
b_n = (2/T) * ∫[0,T] [3sin(2πt/T) + 2cos(4πt/T)] * sin(2πnt/T) dt$

##### Una técnica común para analizar señales no periódicas es la Transformada de Fourier de Tiempo Continuo (FTTC) o la Transformada de Fourier de Tiempo Discreto (FTTD), dependiendo si la señal es continua o discreta en el tiempo. Estas transformadas te permiten descomponer una señal en sus componentes de frecuencia.

$f(t) = e^(-at) * u(t)$
##### Donde a es una constante y $u(t)$ es la función escalón unitario.
##### Para calcular la Transformada de Fourier
$F(w) = ∫[-∞,+∞] f(t) * e^(-jwt) dt$
##### Donde F(w) es la Transformada de Fourier de $f(t)$, j es la unidad imaginaria y w es la frecuencia.
$F(w) = ∫[0,+∞] e^(-at) * e^(-jwt) dt$
##### El cálculo de la Transformada de Fourier de una señal no periódica implica integración en el dominio del tiempo y puede ser más complejo que en el caso de señales periódicas.
